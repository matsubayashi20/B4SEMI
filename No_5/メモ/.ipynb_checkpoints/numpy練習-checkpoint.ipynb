```python
def matrix_inverse(A):
     # 入力された行列のサイズ確認
    A_column = len(A)
    A_row = len(A[0])

    # 出力結果用の変数
    result = [[0] * A_row for i in range(A_column)]
    
    # 正方行列でないならFalseを返す
    if A_column != A_row:
        return False

    # 単位行列の生成
    for i in range(A_column):
        for j in range(A_column):
            if i==j:
                result[i][j] =1
            else:
                result[i][j] =0

    # ガウスの消去法による逆行列の計算(拡大行列[A|I]のi行からk行を足し引きしていく)
    tmp = A
    for i in range(A_column): 
        for k in range(A_column):
            # 同じ行のときスキップする
            if i==k:
                continue

            # 行列Aの対角成分が0ならその行より下の行と入れ替え
            if tmp[i][i] == 0:
                for m in range(i+1,A_column):
                    if tmp[m][i]!=0:
                        for n in range(A_column):
                            tmp_tmp = tmp[m][n]
                            tmp[m][n] = tmp[i][n]
                            tmp[i][n] = tmp_tmp
                            result_tmp = result[m][n]
                            result[m][n] = result[i][n]
                            result[i][n] = result_tmp
                        break

            #入れ替えの結果対角成分が0のままならFalseを返す(ランク落ちしている状態)
            if tmp[i][i] == 0:
                return(False)
            
            # 行列Aの対角成分以外の対角成分と同じ列の成分が0になるように足し引きを行う（変数消去）
            w = tmp[k][i]/tmp[i][i]
            for j in range(i,A_column):
                tmp[k][j]=tmp[k][j]-w*tmp[i][j]
            for j in range(A_column):
                result[k][j] = result[k][j]-w*result[i][j]

    # 行列Aの対角成分が全て1になるように行ごとに拡大行列を割る（拡大行列の右側が逆行列になる）
    for i in range(A_column):
        for k in range(A_column):
            result[i][k] = result[i][k]/tmp[i][i]
    return result

def matrix_transition(A):
    # 結果を入れる変数を用意
    result = []
    
    # 入力された行列のサイズを確認
    A_column = len(A)
    A_row = len(A[0])

    # 転置を行う（エルミート転置）
    for i in range(A_row):
        tmp = []
        for j in range(A_column):
            tmp.append(A[j][i].conjugate())
        result.append(tmp)
    return result
```

```python
import numpy as np
import math

def unitary_matrix(self, vector):
    """
    1. 逆行列とエルミート転置が一致　採用
    2. 行列自身とエルミート転置の積が単位行列

    第三回の荒井のコードを流用
    """
    # 逆行列計算
    inv_vec = matrix_inverse(vector)
    # エルミート転置計算
    her_vec = matrix_transition(vector)

    # 一致するか判別
    if inv_vec == her_vec:
        return True
    else:
        return False
```
