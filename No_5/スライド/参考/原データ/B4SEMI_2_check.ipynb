## 線形独立と線形従属
ベクトル空間 $V$ の $N$ 個の要素 $u_1,u_2,\dots,u_N\in{V}$，<br>スカラ $c_i,i=1,\dots,N$ に対して，<br><br>
$c_1u_1+c_2u_2+\dots+c_Nu_N= 0$ <br><br>
となる必要十分条件が $c_1=c_2=\dots c_N=0$ であるとき，$u_1,u_2,\dots,u_N$ は線形独立といい，そうでない場合は一次従属であると言う．
    

## 1. $A\in{\mathbb{C}^{M \times 1}}$ が$M\times N$行列の列空間に含まれるか判定する関数の仕様①

- ランク $\Longleftrightarrow$ 線形独立な列ベクトル又は行ベクトルの最大本数
- $M\times 1$ が$M\times N$ 行列の列空間に含まれる $\Longleftrightarrow$ $M \times N$行列の線形独立な<br>列ベクトルが$A$の 0 でない要素以上有る
<br>
以上の2点から，$M\times N$行列を$B$ とすると $Rank(B)\ge A$の非零の要素数　<br>の時，行列$A$ は$M\times N$ の列空間には含まれる．



### <font color="DarkOrange"> 上のセルに対するコメント</font>
- 2個目がなんか怪しい \
  $M\times 1$ が$M\times N$ 行列の列空間に含まれる $\Longleftrightarrow$ $M \times N$行列の線形独立な列ベクトルが$A$の 0 でない要素以上有る 
  - 最初の部分は「$A$ が$M\times N$ 行列の列空間に含まれる」とした方がいいかも．「$M\times 1$」だけでは意味が通らない．
  - 「$M \times N$行列の線形独立な列ベクトル」は「$M \times N$行列の線形独立な列ベクトルの個数」とし，\
      「$A$の 0 でない要素」は「$A$の 0 でない要素の数」とした方がより意味が明確になると思う．
  - 以上有る，というので間違いではないと思うけど，もういっそ「...列ベクトルの個数≧...要素の数」とした方がわかりやすいと個人的には思う．
- 最後の部分，列空間「には」というのはどういう意味だろうか．特に何もなければ，「列空間に」でいいと思う．

## 1. $A\in{\mathbb{C}^{M \times 1}}$ が$M\times N$行列の列空間に含まれるか判定する関数の仕様②
### 実装手順
- 与えられた$M\times N$行列の行基本変形を繰り返し，階段行列を作る．

- そのランクが$A$ の非零の要素数よりも大きい場合，True を返し，それ以外は False を返す．

### <font color="DarkOrange"> 上のセルに対するコメント</font>
1点だけ．\
3つ目の項目：そのランクが$A$ の非零の要素数よりも「大きい」場合，True を返し，それ以外は False を返す．\
とあるけど，仕様上「非零の要素数と同じ」場合にもTrueにならないとまずいはず．

## 2.$M\times N$ の行列の列ベクトルが線型独立かどうか判定する関数の仕様
ランクは 1 次独立な列ベクトルの最大本数であるため，
$A\in{\mathbb{C}^{M \times N}}$ について，$Rank(A)=N$ であるとき，$A$の列ベクトルは線形独立である．
<br>
### 実装手順
- 入力行列の行数と列数をチェックし，$M<N$の時ランクが$N$にならないためFalseを返す．
- 与えられた行列の行基本変形を繰り返し，階段行列を作る．
- 階段行列のすべて0でない行の数をカウントし，ランクを求める．
- $Rank(A)=N$の場合はTrueを返す．一致しない場合はFalseを返す．

### <font color="DarkOrange"> 上のセルに対するコメント</font>
$A$の列ベクトルは「全て」線形独立であるの方が正確じゃないかな\
各列ベクトルそれぞれに対して線形独立かどうかは判定可能だし．

## 2.1 $M\times N$行列の列ベクトルが線形独立の場合①
- $M=N=3$の正方行列<br>
$\left(\begin{array}{cc} 1 & 1 & -3 \\ -1 & 3 & 5 \\ 0 & 3 & 6\\ \end{array} \right) \to \left(\begin{array}{cc} 1 & 1 & -3 \\ 0 & 2 & 1 \\ 0 & 0 & \frac{9}{2} \\ \end{array} \right) $<br>$\Longleftrightarrow rank=N=3 \Longleftrightarrow $線形独立 <br>
<br>
予期される出力はTrue

### <font color="DarkOrange"> 上のセルに対するコメント</font>
